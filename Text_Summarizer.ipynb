```python
import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
WHITESPACE_HANDLER = lambda k: re.sub('\s+', ' ', re.sub('\n+', ' ', k.strip()))
```

```python
# Prompting the user for the article text
article_text = input("Please enter the article text you want to summarize: ")

tokenizer = AutoTokenizer.from_pretrained("padmesh2004/text_summarizer")
model = AutoModelForSeq2SeqLM.from_pretrained("padmesh2004/text_summarizer")
```

```python
# Tokenize the article text
input_ids = tokenizer(
    [WHITESPACE_HANDLER(article_text)],
    return_tensors="pt",
    padding="max_length",
    truncation=True,
    max_length=512
)["input_ids"]
```

```python
# Generate the summary
output_ids = model.generate(
    input_ids=input_ids,
    max_length=150,
    num_beams=4,
    early_stopping=True,
)[0]
```

```python
# Decode and print the summary
summary = tokenizer.decode(
    output_ids,
    skip_special_tokens=True,
    clean_up_tokenization_spaces=False
)
```

```python
print(summary)
```
